```python
# Creacion de arreglo con ceros
zeros=np.zeros(3)
print(zeros)
```

```python
# Creacion de arreglo con unos
ones=np.ones(3)
print(ones)
```

```python
# Creacion de matriz con ceros
zerosMatrix=np.zeros((1024,3))
print(zerosMatrix)
```

```python
# Creacion de matriz con unos
onesMatrix=np.ones((1024,3))
print(onesMatrix)
```

```python
# Empty
# Valores aleatorios muy pequenios
# Creacion de matriz con empty
randomis=np.empty((1024,3))
print(randomis)
```

```python
# array en arange
arr=np.arange(10)
print(arr)
```

```python
# array en arange con parametros
# Inicia en 0, menor al 20, con aumento de 3
arr=np.arange(0,20,2)
print(arr)
```

```python
# linspace
# Create an array that are spaced linearly in an interval.
arr=np.linspace(0,15,num=2)
print(arr)
```

```python
# ordenar un arreglo
arr=np.array([4,2,5,6,3,2,4,2,7,8,9,4,3,3])
np.sort(arr)
```
